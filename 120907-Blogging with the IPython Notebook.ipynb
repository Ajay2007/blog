## Converting your notebook to html with nbconvert

The first thing you will need is our [nbconvert](https://github.com/ipython/nbconvert) tool that converts notebooks across formats.  The README file in the repo contains the requirements for nbconvert (basically [pandoc](http://johnmacfarlane.net/pandoc), [docutils from SVN](http://docutils.svn.sourceforge.net/viewvc/docutils/trunk/docutils/?view=tar) and [pygments](http://pygments.org)).

Once you have nbconvert installed, you can convert your notebook to html with

    nbconvert -f html your_notebook.ipynb

This will leave a file named `your_notebook.html`, as well as a directory `your_notebook_files` with any ancillary files that it may need to create (for referenced images):

```
(master)dreamweaver[notebook]> ls
notebook_blogging.ipynb
(master)dreamweaver[notebook]> nbconvert.py -f html notebook_blogging.ipynb 
(master)dreamweaver[notebook]> ls
notebook_blogging_files/  notebook_blogging.html  notebook_blogging.ipynb
```

## Configuring your Blogger blog to accept notebooks

The notebook uses a lot of custom CSS, and you will need all this CSS in your blog's configuration for your notebook-based posts to display correctly.  [This site](http://smallbusiness.chron.com/customize-css-blogspot-29958.html) has detailed instructions on how to achieve this, but quickly they can be summarized as:

1. Once authenticated, go to your blog's overview page by clicking on its title.
2. Click on templates (left column) and customize using the Advanced options.
3. Scroll down the middle column until you see an "Add CSS" option.
4. Copy the IPython CSS in there.
5. Click the "Apply to Blog" button at the top right.

To get the IPython CSS, simply open the HTML file that nbconvert created in a text editor, and copy all the text starting with

    <style type=text/css>
    
until the last `style` block, which appears right before the first `<script>` tag.

That's it, and you shouldn't need to do anything else as long as the CSS we use in the notebooks doesn't drastically change.  This customization of your blog needs to be done only once.

While you are at it, I recommend you change the width of your blog so that cells have enough space for clean display; in experimenting I found out that the default template was too narrow to properly display code cells, producing a lot of text wrapping that impaired readability.

*Note:* if you change blog templates, Blogger does destroy your custom CSS, so you may need to repeat the above steps in that case.

## Adding the actual posts

Now, whenever you want to write a new post as a notebook, simply convert the `.ipynb` file to html, open it in an editor, and for each of the following blocks of text:

1. (Only needed if you use math) The scripts that initialize mathjax: copy from the `<script>` right after the last piece of CSS to the last `</script>` tag (it's only ~ 20 lines of Javascript).

2. All the text from immediately between the `<body>` and `</body>` tags.  This is the body of your post.

Copy them to the clipboard, then go to the 'raw html' view of the post, remove anything Blogger may have put there by default, and paste them.  You should also click on the 'options' tab (right hand side) and select both `Show HTML literally` and `Use <br> tag`, else your paragraph breaks will look all wrong.

That's it!

In the future I'm sure we'll streamline this a bit, and will add an option to generate blogger-friendly html with the CSS in one file and just the pure html body in another so this process is more convenient.  But for now, the manual steps are pretty straightforward.

## Including code examples from other languages

Using our various script cell magics, it's easy to include code in a variety of other languages

````python
%%ruby
puts "Hello from Ruby #{RUBY_VERSION}"
````

````python
%%bash
echo "hello from $BASH"
````

And tools like the Octave and R magics let you interface with entire computational systems directly from the notebook; this is the Octave magic for which [our example notebook contains more details](http://nbviewer.ipython.org/urls/raw.github.com/ipython/ipython/d835d46dcc50043971b4a9915398bad1b5d63648/docs/examples/notebooks/octavemagic_extension.ipynb):

````python
%load_ext octavemagic
````

````python
%%octave -s 500,500

# butterworth filter, order 2, cutoff pi/2 radians
b = [0.292893218813452  0.585786437626905  0.292893218813452];
a = [1  0  0.171572875253810];
freqz(b, a, 32);
````

The [rmagic extension](http://nbviewer.ipython.org/urls/raw.github.com/ipython/ipython/3607712653c66d63e0d7f13f073bde8c0f209ba8/docs/examples/notebooks/rmagic_extension.ipynb) does a similar job, letting you call R directly from the notebook, passing variables back and forth between Python and R.

````python
%load_ext rmagic 
````

Start by creating some data in Python

````python
X = np.array([0,1,2,3,4])
Y = np.array([3,5,4,6,7])
````

Which can then be manipulated in R, with results available back in Python (in `XYcoef`):

````python
%%R -i X,Y -o XYcoef
XYlm = lm(Y~X)
XYcoef = coef(XYlm)
print(summary(XYlm))
par(mfrow=c(2,2))
plot(XYlm)
````

````python
XYcoef
````

And finally, in the same spirit, the [cython magic extension](http://nbviewer.ipython.org/urls/raw.github.com/ipython/ipython/3607712653c66d63e0d7f13f073bde8c0f209ba8/docs/examples/notebooks/cython_extension.ipynb) lets you call Cython code directly from the notebook:

````python
%load_ext cythonmagic
````

````python
%%cython -lm
from libc.math cimport sin
print 'sin(1)=', sin(1)
````
